## The Multilayer Perceptron: A step by step
As the name implies, a multilayer perceptron (MLP) is a set of perceptrons arranged into layers. The goal here is to have a learner that:
* works for both regression and classification
* is a universal function approximator

To do this a few changes will need to be made to the perceptron:
* The threshold will need to be replaced with an activation function. In this notebook, we'll be using the sigmoid function: $$g(z) = \frac{1}{1 + e^{-z}}$$

* A differentiable loss function will need to be used. We'll use the $L_2$ loss function in this notebook.

With the above in mind, we can obtain an MLP learner by having one or more layers between the input and output. The figure below depicts an MLP with one hidden layer. This is the quintessential neural network (NN); it is fully connected and feed-forward.

Let's implement this MLP network from scratch in a step by step manner. After that, we'll put everything together in a class. 

### Determining the shape of the MLP network
In MLP:
* the number of features in the dataset determines how many nodes in the input layer
* the hidden layer can have as many nodes as needed. So we'll need a variable(say `nHidden`) to hold that number
* the output layer can have more than one node. This works well for multi-class classification.

These numbers determine the shape of the network. Here are some example values:

```python
nFeatures = 2 # Used to determine how many nodes in the input layer
nHidden = 2   # Used to determine how many nodes in the hidden layer
nOutput = 1   # Used to determine how many nodes in the output layer
```

Training an MLP model involves two main operations:

### A) Moving forward
Moving forward means taking an input and passing it across the network until the output is reached and calculated. 

```python
def forward(inputs):
    """Assumes that inputs cotains the bias column"""

    # Calculaing the activations of the hidden nodes
    a = np.dot(inputs, w);
    a = 1.0/(1.0 + np.exp(-a))
    
    # Adding bias to the hidden nodes
    a = np.concatenate([np.ones((len(inputs), 1)), a], axis=1)

    # Calculating the output
    y = np.dot(a, v).squeeze()

    return a, (1.0 / (1.0 + np.exp(-y)))
```

### B) Moving backward
Moving backward means updating the weights of the network (where the actual learning happens) such that the loss function at the output layer is minimized. Because the we only know the error at the output layer, we need to propagate there error backward in order to update these weights in a way that minimize that error. We call this **backpropagation**.

First we calculate a few error quantities at both the output and hidden layers using the following formulas (in matrix notation):

$$\delta_o = (\mathbf{y} - \mathbf{t}) \mathbf{y} (1 - \mathbf{y})$$

$$\delta_h = \mathbf{a} (1 - \mathbf{a}) \delta_o \mathbf{v}^T$$

where $\mathbf{t}$ is the actual target output, $\mathbf{y}$ is the predicted output, $\mathbf{a}$ are the activations of the hidden layers as the figure above shows.

We use these perceived errors to update the weights:

$$ \mathbf{w} = \mathbf{w} - \eta \mathbf{X} \delta_h^*$$
$$ \mathbf{v} = \mathbf{v} - \eta \mathbf{a} \delta_a$$

where $\delta_h^*$ is the same as $\delta_h$ except that the first bias column is removed. Again this is the backpropagation and here is a function to implement it.

```python
def backward(X, t, a, y, eta):
    global w, v
    a, y = forward(X)
    error = 0.5* np.sum((y - t)**2)
    errors.append(error) 

    # Calculate the "perceived errors"
    delta_o = ((y - t) * y *(1.0 - y)).reshape(-1, 1)
    delta_h = a *(1.0 - a) * (np.dot(delta_o, np.transpose(v)))

    # Update weights
    w -= eta * (np.dot(np.transpose(X),delta_h[:, 1:]))
    v -= eta * (np.dot(np.transpose(a), delta_o))
```

### Fitting the data
We can now use the above forward and backward functions to fit (train) the MLP network to the data. This is where the "learning" happens. We start by initializing the weights. As the above figure shows, we have two sets of weights:
* The weights $w_{ij}$ between the input layer and the hidden layer. 
* The weights $v_{jk}$ between the hidden layer and the output layer.

These weights will need to be initialized properly. A good rule of thump is to initialize them to a random number between $[-1/\sqrt{n}, +1/\sqrt{n}]$ where $n$ is the number of nodes in layer preceding these weights.

```python
w = None
v = None
def fit(X, y, eta=.2, epochs=1000):
    # Initialze weights
    global w, v
    w = (np.random.rand(nFeatures + 1, nHidden)-0.5) * 2 / np.sqrt(nFeatures)
    v = (np.random.rand(nHidden + 1, nOutput)-0.5) *2/np.sqrt(nHidden)
    
    # Adding bias to the input nodes
    X = np.concatenate((np.ones((len(X),1)), X.reshape(len(X), -1) if X.ndim == 1 else X), axis=1)
    for n in range(epochs):
        backward(X, y, *forward(X), eta)
```

### Making predictions
Having fit the data, we now can use the trained network to make predictions. Here is a function that performs classification. This function uses `.05` as a threshold function: if $y > .5$, return 1 else return 0.

```python
def predict(X_unseen):
    if X_unseen.ndim == 1:
        X_unseen = X_unseen.reshape(1, len(unseen))
    
    # Adding bias to unseen
    X_unseen = np.concatenate((np.ones((len(X_unseen),1)), X_unseen), axis=1)
    h, y = forward(X_unseen)
        
    return np.where(y > 0.5, 1, 0)
```

## Putting everything together: The MLP class

Let's put all this together into a class. This class will be able to carry out both binary and multi-class classifications. For the latter (multi-class classification), two changes are needed:
* The output column will need to be one-hot encoded. 
* The predict function will be changed to return the class (label) with the large output probability

```python
class MultilayerPerceptron:
    def __init__(self, nHidden=None, eta=.2, epochs=1000):
        """
        nHidden: determines how many nodes in the hidden layer
        eta: the learning rate
        epochs: how many times the whole dataset is gone over
        """
        self.nHidden = nHidden
        self.eta = eta
        self.epochs = epochs
        
    def fit(self, X, y):
        dataset = my.to_dataframe(X, y)
        self.labels = np.unique(dataset.iloc[:, -1])
        self.inputs = dataset.values[:, :-1]
        self.target = dataset.values[:, -1:]
        
        """
        nOutput: determines how many nodes in the output layer. For binary classification
                 nOutput = 1. For multi-class classification, one-hot-encoding is used
                 and nOutput = the number unique classes (labels) in the output column.
        """
        self.nOutput = 1
        
        """ One-hot-encoding to support multi-class classification """
        if len(self.labels) > 2:
            self.nOutput = len(self.labels)
            
            self.target = pd.DataFrame({
                f"y{i + 1}": dataset.iloc[:, -1].apply(
                    lambda x: 1 if x == self.labels[i] else 0
                ) for i in range(self.nOutput)
            }).values
            
        self.nFeatures = self.inputs.shape[1]
        self.nHidden = self.nFeatures if self.nHidden == None else self.nHidden
        self.w = []
        
        self.errors = []
        self.__initialize_weights()
        
        for n in range(self.epochs):
            self.backward(*self.forward(self.inputs))
            
        return self
        
    def activation_f(self, h):
        """
        The activation (sigmoid) function
        """
        return 1.0/(1.0 + np.exp(-h))
    
    def forward(self, X):
        if len(self.w) == 0:
            raise "Need to fit the data first"
        
        X = np.concatenate([np.ones((len(X),1)), X],axis=1)
        
        # Hidden layer 1
        self.a = []
        a0 = self.activation_f(np.dot(X, self.w[0]));
        
        # Adding bias to the hidden nodes
        a0 = np.concatenate([np.ones((len(a0), 1)), a0], axis=1)
        self.a.append(a0)

        return X, self.activation_f(np.dot(self.a[0], self.w[1]))
    
    def backward(self, X, y):
        self.errors.append(0.5* np.sum((y - self.target)**2)) 

        # Calculate the "perceived errors"
        delta_o = (y - self.target) * y *(1.0 - y) 
        delta_h = self.a[0] *(1.0 - self.a[0]) * (np.dot(delta_o, self.w[1].T))

        # Update weights
        self.w[0] -= self.eta * (np.dot(X.T, delta_h[:,1:]))
        self.w[1] -= self.eta * (np.dot(self.a[0].T, delta_o))
        
    def __initialize_weights(self):
        self.w.append((np.random.rand(self.nFeatures + 1, self.nHidden) - 0.5) * 2 / np.sqrt(self.nFeatures))
        self.w.append((np.random.rand(self.nHidden + 1, self.nOutput)-0.5) * 2 / np.sqrt(self.nHidden))
    
    def predict(self, X_unseen):
        if X_unseen.ndim == 1:
            X_unseen = X_unseen.reshape(1, len(X_unseen))

        X, y = self.forward(X_unseen)
        if self.target.shape[1] == 1:
            return np.where(y > 0.5, 1, 0)
        else: # For multi-class classification
            return np.argmax(y,1)
```

to scale the input values. Normalization (min-max scaling) uses the formula:

$$ x_{new} = \frac{x - x_{min}}{x_{max} - x_{min}}$$

to scale the input values.

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

Keep in mind that this dataset has been processed, and each image has been reduced to 8 by 8 pixels. Every pixel is represented in the input by a color value between 0 and 16. The `digits` dataset has two instance variables: `images` for the input and `target` for the output labels. Let's take a look at some of these digits:

```python
fig = plt.figure(1, figsize=(16, 7))
for i in range(0, 10):
    ax = fig.add_subplot(1, 10, i + 1)
    ax.imshow(digits.images[i], cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title(f"Digit: {digits.target[i]}")
```

We start by reshaping each image to a row with 64 features (one per pixel). We also need to normalize the color values.

```python
X = digits.images.reshape(1797, 64)
X = X / 16 # Input normalization
y = digits.target.reshape(len(digits.target), 1)
df = pd.DataFrame(np.concatenate((X, y), axis=1))
df
```

Let's train an MLP to classify this data:

```python
X, y, features, target = my.from_dataframe(df)
X_train, X_test, y_train, y_test = my.train_test_split(X, y, test_size=.2)
    
mlp = MultilayerPerceptron( nHidden=15, epochs=1000, eta=.01)
mlp.fit(X_train, y_train)

cm = my.confusion_matrix(y_train, mlp.predict(X_train))
accuracy = np.trace(cm) / np.sum(cm)
print(cm)
print('Training accuracy: ', accuracy)

cm = my.confusion_matrix(y_test, mlp.predict(X_test))
accuracy = np.trace(cm) / np.sum(cm)
print(cm)
print('Testing accuracy: ', accuracy)
```

Here are what the learning curves look like:

```python
epochs = [1, 10, 50, 100, 150, 200, 300, 400, 500, 700, 900, 1000]
train_err = []
test_err = []
for n in epochs:
    mlp = MultilayerPerceptron(nHidden=15, epochs=n, eta=.01).fit(X_train, y_train)
    
    cm = my.confusion_matrix(y_train, mlp.predict(X_train))
    train_err.append(1 - np.trace(cm) / np.sum(cm))

    cm = my.confusion_matrix(y_test, mlp.predict(X_test))
    test_err.append(1 - np.trace(cm) / np.sum(cm))
 
fig = plt.figure(figsize=(16,5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(epochs, train_err, label="Training")
ax.plot(epochs, test_err, label="Testing")
ax.set_xlabel("epochs")
ax.set_ylabel("error")
ax.legend()
ax.set_title("MLP learning curves - using epochs")
plt.show()
```

## Using SciKit Learn's MLP
SciKit Learn has two MLP implementations: `sklearn.neural_network.MLPClassifier` for classification and `sklearn.neural_network.MLPRegressor` for regression.

Let's see an example with with two kinds of scaling features: normalization and standardization. To do that, we use `StandardScaler` (for standardization) and `MinMaxScaler` (for normalization). Both of these scalers are transformers implemented by Scikit Learn's `preprocessing` package. They both provide a `fit_transform` method to do the actual scaling. Let's try both of them side by side.

For this example we will be using the the [Wine Dataset](https://archive.ics.uci.edu/ml/datasets/wine). Let's read it in.

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', header=None)

df.columns = [
    'y', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 
    'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 
    'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline'
]
```

As you see, the first column is the output column. Let's move it to the end.

```python
df = df[[*df.columns[1:], df.columns[0]]]
df
```

Let's split it into training and testing portions. This time we are passing Pandas' dataframes to the `train_test_split()` function.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:-1], df.iloc[:, -1], test_size=0.2)
```

We now define the `MinMaxScaler` and `StandardScaler`:

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

X, y = df.iloc[:,:-1], df.iloc[:, -1]

std_scaler, mmx_scaler = StandardScaler(), MinMaxScaler()
X_standarized, X_normalized = std_scaler.fit_transform(X), mmx_scaler.fit_transform(X)
```

Let's print the original input data along with the scaled ones:

```python
X.to_numpy()[:3,], X_standarized[:3,], X_normalized[:3,]
```

### Using the standardized data

We are now ready to split the data and train an MLP model. First we use the standardized data. Let's split the data.

```python
from sklearn.neural_network import MLPClassifier
X_train, X_test, y_train, y_test = train_test_split(X_standarized, y, test_size=0.2)
```

And train the MLP model and plot the confusion matrix:

```python
from sklearn.metrics import ConfusionMatrixDisplay
mlp = MLPClassifier(random_state=1, max_iter=700).fit(X_train, y_train)

ConfusionMatrixDisplay.from_estimator(mlp, X_test, y_test)
```

We can also use print the classification report which displays many metrics including recall and precision

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, mlp.predict(X_test)))
```

### Using the standardized data

Hand here is the same using the normalized data.

```python
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2)
mlp = MLPClassifier(random_state=1, max_iter=1000).fit(X_train, y_train)

ConfusionMatrixDisplay.from_estimator(mlp, X_test, y_test)

print(classification_report(y_test, mlp.predict(X_test)))
```
