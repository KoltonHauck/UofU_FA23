# 05: Nearest neighbors (NN) classifiers

```python
%matplotlib inline

import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mylib as my
```

## Nearest Neighbor (NN)
Given and dataset $\boldsymbol{D} = [\boldsymbol{X}\ \ \boldsymbol{y}]$ and an unseen example $\boldsymbol{x}$, this classifier:
* calculates the distances between $\boldsymbol{x}$ and every example in $\boldsymbol{D}$. Using the Euclidean distance, we have:
    $$
      distance(\boldsymbol{x^{(a)}}, \boldsymbol{x^{(b)}}) = \Vert\boldsymbol{x^{(a)}} - \boldsymbol{x^{(b)}}\Vert_2 = \sqrt{\sum_{i = 1}^{m} (x_i^{(a)} - x_i^{(b)})^2}
    $$
* finds the example that is nearest to X and return its $y$ 

Here is a class implementing this classifier:

```python
class NearestNeighborClassifier:
    def calc_distance(self, x, y):
        distance = 0.0
        for i in np.arange(len(x)):
            distance += (x[i] - y[i])**2
                
        return math.sqrt(distance)
        
    def fit(self, X, y):
        self.X = X
        self.y = y
        
        return self
        
    def predict(self, X_unseen):
        if X_unseen.ndim == 1:
            N = len(self.X)
            distances = np.full(N, 0.0)
            for i, ex in enumerate(self.X):
                distances[i] = self.calc_distance(X_unseen, ex)

            return self.y[distances.argmin()]
        else:
            return np.array([self.predict(ex) for ex in X_unseen])
```

## An example
Let's quickly test this. Here is a sample dataset with three classes:

```python
basis = np.array([[3, 14],[4, 4],[10, 9]])
a_ = np.concatenate((
        basis[0,0] + np.random.normal(loc=1, scale=2.5, size=(50, 1)), 
        basis[0, 1] + np.random.normal(loc=1, scale=2.6, size=(50, 1)), 
        np.full(50, 0).reshape(50,1)), axis= 1)

b_ = np.concatenate((
        basis[1,0] + np.random.normal(loc=.8, scale=2.5, size=(50, 1)), 
        basis[1, 1] + np.random.normal(loc=.8, scale=2.4, size=(50, 1)), 
        np.full(50, 1).reshape(50,1)), axis= 1)

c_ = np.concatenate((
        basis[2,0] + np.random.normal(loc=.5, scale=2.5, size=(50, 1)), 
        basis[2, 1] + np.random.normal(loc=.5, scale=2.6, size=(50, 1)), 
        np.full(50, 2).reshape(50,1)), axis= 1)
```

which looks like this:

```python
plt.figure(figsize=(16,6))
plt.scatter(a_[:, 0], a_[:, 1], color='red', marker='x', label='A')
plt.scatter(b_[:, 0], b_[:, 1], color='blue', marker='+', label='B')
plt.scatter(c_[:, 0], c_[:, 1], color='green', marker='^', label='C')

plt.legend(loc='best')
plt.title('Samle three classes data')
```

Let's put this data into a dataset and split it into training and test sets.

```python
data =np.concatenate([a_, b_, c_], axis=0)
X = data[:, :2]
y = data[:, 2]

X_train, X_test, y_train, y_test = my.train_test_split(X, y, test_size=.33)
```

Let's train this classifier and print its output on the training set side by side with the actual training labels.

```python
nn = NearestNeighborClassifier().fit(X_train, y_train)

results = pd.DataFrame(columns=["Actual", "Predicted"])

results["Actual"] = y_train.flatten()
results["Predicted"] = nn.predict(X_train)
results.head(10)
```

Here is the corresponding confusion matrix and accuracy.

```python
cm = my.confusion_matrix(y_train, nn.predict(X_train))
accuracy = np.trace(cm) / np.sum(cm)

print(cm)
print('Train accuracy: ', accuracy)
```

This is a perfect accuracy which corresponds to a training error of $\boldsymbol{E}_{train} = 0$. This is not necessarily a good thing. **Why?**

Let's see how this trained model performs on the unseen test set. 

```python
results = pd.DataFrame(columns=["Actual", "Predicted"])

results["Actual"] = y_test.flatten()
results["Predicted"] = nn.predict(X_test)
results.head(10)
```

Here is the corresponding confusion matrix and accuracy.

```python
cm = my.confusion_matrix(y_test, nn.predict(X_test))
accuracy = np.trace(cm) / np.sum(cm)

print(cm)
print('Test accuracy: ', accuracy)
```

Here are the decision regions of this trained model with both the training examples and the test examples side by side. **Uncomment the `%%timeit` command to see how long it takes to produce these plots and get a feeling of how computationally expensive the NN classifier can be.** Keep in mind that `%%timeit` will run your code multiple times to get a more accurate measure of how long it took. So it may take a while to finish.

```python
#%%timeit
fig = plt.figure(figsize=(16,5))
ax = fig.add_subplot(1,2,1)
my.plot_decision_regions(X_train, y_train, nn, title="Training data", ax=ax)
ax = fig.add_subplot(1,2,2)
my.plot_decision_regions(X_test, y_test, nn, title="Test data", ax=ax)
```

## A more efficient implementation
The way class `NearestNeighborClassifier` calculated the distances can be improved by taking advantage of NumPy vectorization. Here is a shorter and improved version. **Do you see the difference?**

```python
class NearestNeighborClassifier:
    def fit(self, X, y):
        self.X = X
        self.y = y
        
        return self
    
    def predict(self, X_unseen):
        if X_unseen.ndim == 1:
            distances = np.sqrt(np.sum((self.X - X_unseen)**2, axis=1))

            return self.y[distances.argmin()]
        else:
            return np.array([self.predict(ex) for ex in X_unseen])
```

Using the above example, to test this new class gives the following for the training set 

```python
nn = NearestNeighborClassifier().fit(X_train, y_train)

results = pd.DataFrame(columns=["Actual", "Predicted"])

results["Actual"] = y_train
results["Predicted"] = nn.predict(X_train)
print(results.head(10))

cm = my.confusion_matrix(y_train, nn.predict(X_train))
accuracy = np.trace(cm) / np.sum(cm)

print(cm)
print('Train accuracy: ', accuracy)
```

and the following for the test set:

```python
results = pd.DataFrame(columns=["Actual", "Predicted"])

results["Actual"] = y_test
results["Predicted"] = nn.predict(X_test)
print(results.head(10))

cm = my.confusion_matrix(y_test, nn.predict(X_test))
accuracy = np.trace(cm) / np.sum(cm)

print(cm)
print('Test accuracy: ', accuracy)
```

This is no longer a perfect accuracy meaning that test error of $\boldsymbol{E}_{test} > 0$.

Here are again the decision regions of this trained model with both the training examples and the test examples side by side. **Uncomment the `%%timeit` command to see how long it takes to produce these plots using the improved NN classifier and compare that to the performance of the first implementation.**

```python
#%%timeit
fig = plt.figure(figsize=(16,5))
ax = fig.add_subplot(1,2,1)
my.plot_decision_regions(X_train, y_train, nn, title="Training data", resolution=.1, ax=ax)
ax = fig.add_subplot(1,2,2)
my.plot_decision_regions(X_test, y_test, nn, title="Test data", resolution=.1, ax=ax)
```

## Using `scikit-learn` estimators/predictors

Scikit-learn comes with all kinds of supervised machine learning models. They are both **estimators** and **predictors**. They can also be broken into two groups: classifiers and regressors. A model of these is typically implemented using a class. The constructor of that class is where hyperparameters are defined and given sensible default values. It also consists of multiple methods. Two methods in particular are implemented by all of these classifiers and regressors: 

* `fit(X, y)` which is required by the estimator interface. It takes a training dataset as a parameter. This is where "learning" happens.
* `predict(X)` which is required by the predictor interface to makes predictions for given unseen data.

Having these two methods give these models a consistent interface and makes it easy to use all sort of models once you know how to use one. As an example, we'll train a Nearest Neighbor (NN) classifier to predict the flower type given the above data.

```python
from sklearn.neighbors import KNeighborsClassifier
nn = KNeighborsClassifier(n_neighbors=1)
nn.fit(X_train, y_train)
```

Let's plot the training confusion matrix and print its accuracy. We'll create a function for doing so and add it to our own `mylib` package. Scikit-learn provides functions under the `metrics` package for that.

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score

def plot_confusion_matrix_with_accuracy(classifier, X_test, y_test, ax=None):
    actual, predicted = y_test, classifier.predict(X_test)
    cm = confusion_matrix(actual, predicted)
    ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, ax=ax)
    print(f"Accuracy: {accuracy_score(actual, predicted)}")
    
plot_confusion_matrix_with_accuracy(nn, X_train, y_train)
```

1.0 accuracy score (or 100% accuracy) wow! Perhaps, however, we are overfitting. 

## K-fold cross-validation
Let's convert the above classifier to a K-Nearest Neighbors model. If you call the function:

```python
nn.get_params()
```

You will get the list of hyper-parameters the model has. One of these parameters is `n_neighbors` which is the `K` in K-NearestNeighbor. Since we don't really know the best value for this parameter, we use K-fold cross-validation to find that out. Sciket-learn's `model_selection` package provides multiple functions for implementing cross-validation. Once such function is simply called `cross_validate`. 

Here is an example using this function to find out the best value for the `n_neighbors` parameter. It's critical to always remember that in Scikit-learn, cross-validation functions don't use errors or loss functions for scoring. Instead they use utility functions where larger is better. As you see below, we are using the accuracy for scoring instead of the misclassification error.

```python
from sklearn.model_selection import cross_validate
ks, scores =[], []
for k in range(1,16):
    nn = KNeighborsClassifier(n_neighbors=k)
    cv_results = cross_validate(nn, X_train, y_train, scoring='accuracy', cv=10)
    
    np.mean(cv_results['test_score'])
    ks.append(k)
    scores.append(np.mean(cv_results['test_score']))
```

Here is a function to plot the cross-validation curve. In the code above, the `cv=10` parameter means 10-fold cross-validation.

```python
def plot_cv_curve(x, y, xlabel, ylabel, title="", ax=None):
    if ax is None:
        fig, ax = plt.subplots(1, 1, figsize=(16,5))
    ax.plot(x, y)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)

plot_cv_curve(ks, scores, "k", "accuracy", "Tuning parameter k")    
```

Here is the best value for the `n_neighbors` parameter. The `+ 1` is because `np.argmax` returns a 0-based index and our parameter values start at 1.

```python
np.argmax(scores) + 1
```

We now train a KNN model with this best `n_neighbors` value:

```python
knn = KNeighborsClassifier(n_neighbors=np.argmax(scores) + 1)
knn.fit(X_train, y_train)
```

And we can finally test this model using our set-aside test dataset. Let's plot the testing confusion matrix and print its accuracy.

```python
my.plot_confusion_matrix_with_accuracy(knn, X_test, y_test)
```

And here are the decision regions of this `scikit-learn` model. Notice how easy it is to use `scikit-learn` models with our own custom-defined functions.

```python
#%%timeit
fig = plt.figure(figsize=(16,5))
ax = fig.add_subplot(1,2,1)
my.plot_decision_regions(X_train, y_train, knn, title="Training data", resolution=.1, ax=ax)
ax = fig.add_subplot(1,2,2)
my.plot_decision_regions(X_test, y_test, knn, title="Test data", resolution=.1, ax=ax)
```
