## Multinomial naive Bayes classifier
Given an unseen example $\boldsymbol{x}$, we want to predict its corresponding $y$ from a set of possible classes $y_1, y_2, \dots$. We can do so by picking the $y$ such that:
$$ y = \operatorname*{argmax}_{i}\ \ P(y_i | \boldsymbol{x})$$

We can calculate $P(y_i | \boldsymbol{x})$ using the formula:
$$ P(y_i | \boldsymbol{x}) = \alpha\ P(y_i) P(\boldsymbol{x} | y_i) $$

where $\alpha$ is a normalizing factor and $P(y_i)$ is the prior probability which can be estimated from the training dataset.

What makes **naive Bayes** naive is the strong assumption that the features $x_1, x_2, x_3, \cdots, x_m$ are independent from one another. Such an assumption allows us to write:

$$P(\boldsymbol{x} | y_i) = P(x_1, x_2, x_3, \cdots, x_m| y_i) = P(x_1 | y_i)P(x_2 | y_i)P(x_3 | y_i)\cdots P(x_m | y_i)$$
 
The $P(x_j | y_i)$ is the likelihood of $x_j$ given the class $y_i$, which can be estimated from the training set as well.

To see how this classifier works, let's go through a detailed example:

### The problem

Let's begin by introducing the data we will be using, adapted from 'Machine Learning: An Algorithmic Perspective, Second Edition' by Stephen Marsland.

```python
df = pd.DataFrame({
    'deadline': ['Urgent', 'Urgent', 'Near', 'None', 'None', 'None', 'Near', 'Near', 'Near', 'Urgent'],
    'party': ['Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'No'],
    'lazy': ['Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No'],
    'y': ['Party', 'Study', 'Party', 'Party', 'Pub', 'Party', 'Study', 'TV', 'Party', 'Study']
})

df
```

Given this data, what should $y$ be for the following example? 

```python
x = { "deadline": "Near", "party": "No", "lazy": "No" }
```

### How does it work?

Given an unseen example $\boldsymbol{x}$, we want to predict its corresponding $y$ from a set of possible classes (labels) $y_1, y_2, \dots$. We can do so by picking the $y$ such that:
$$ y = \operatorname*{argmax}_{i}\ \ P(y_i | \boldsymbol{x})$$

This basically translates to the following steps:

* For each of the possible classes (labels) $y_i$ in the output column, calculate $P(y_i | \boldsymbol{x})$ using the formula:

    $$ P(y_i | \boldsymbol{x}) = P(y_i) P(x_1 | y_i)P(x_2 | y_i)P(x_3 | y_i)\cdots P(x_m | y_i)$$
 
* Return the class (label) $y_i$ that corresponds to the largest value for $P(y_i | \boldsymbol{x})$

In this example problem we have:

* Four output classes: $y_1 = Party,\ y_2 = Study,\ y_3 = TV,\ and\ y_4 = Pub$ 
* Three input features: deadline, party (whether there is a party or not), lazy(whether you are feeling lazy or not)
* The unseen example $x$ has three values: $x_1$ corresponds to its value for the deadline feature, $x_2$ to its value for the party feature, and $x_3$ to its value for the lazy feature.


### Answer

First we find out the possible output classes $y_i$:

```python
df['y'].unique()
```

For each one of these classes, we need to calculate $p(y_i | \boldsymbol{x})$

1. For the case where $y_i = Party$, we need to calculate $$ P(y_i=Party|\boldsymbol{x}) = P(y_i=Party) * P(x.deadline|y_i=Party) * p(x.party|y_i=Party) * p(x.lazy|y_i=Party)$$



```python
# Start with the reduced sample space - only the examples where y is Party 
sub_df = df[df['y'] == 'Party']
sub_df

p_of_party_given_x = (
    len(sub_df) / len(df) *
    len(sub_df[sub_df['deadline'] == x['deadline']]) / len(sub_df) *
    len(sub_df[sub_df['party'] == x['party']]) / len(sub_df) *
    len(sub_df[sub_df['lazy'] == x['lazy']]) / len(sub_df)
)

p_of_party_given_x
```

2. For the case where $y_i = Study$, we calculate $$ P(y_i=Study|\boldsymbol{x}) = P(y_i=Study) * P(x.deadline|y_i=Study) * p(x.party|y_i=Study) * p(x.lazy|y_i=Study)$$

```python
# Start with the reduced sample space - only the examples where y is Study
sub_df = df[df['y'] == 'Study']
sub_df

p_of_study_given_x = (
    len(sub_df) / len(df) *
    len(sub_df[sub_df['deadline'] == x['deadline']]) / len(sub_df) *
    len(sub_df[sub_df['party'] == x['party']]) / len(sub_df) *
    len(sub_df[sub_df['lazy'] == x['lazy']]) / len(sub_df)
)

p_of_study_given_x
```

3. For the case where $y_i = TV$, we calculate $$ P(y_i=TV|\boldsymbol{x}) = P(y_i=TV) * P(x.deadline|y_i=TV) * p(x.party|y_i=TV) * p(x.lazy|y_i=TV)$$

```python
# Start with the reduced sample space - only the examples where y is TV
sub_df = df[df['y'] == 'TV']
sub_df

p_of_tv_given_x = (
    len(sub_df) / len(df) *
    len(sub_df[sub_df['deadline'] == x['deadline']]) / len(sub_df) *
    len(sub_df[sub_df['party'] == x['party']]) / len(sub_df) *
    len(sub_df[sub_df['lazy'] == x['lazy']]) / len(sub_df)
)

p_of_tv_given_x
```

4. And finally for the case where $y_i = Pub$, we need to calculate $$ P(y_i=Pub|\boldsymbol{x}) = P(y_i=Pub) * P(x.deadline|y_i=Pub) * p(x.party|y_i=Pub) * p(x.lazy|y_i=Pub)$$

```python
sub_df = df[df['y'] == 'Pub']
sub_df

p_of_pub_given_x = (
    len(sub_df) / len(df) *
    len(sub_df[sub_df['deadline'] == x['deadline']]) / len(sub_df) *
    len(sub_df[sub_df['party'] == x['party']]) / len(sub_df) *
    len(sub_df[sub_df['lazy'] == x['lazy']]) / len(sub_df)
)

p_of_pub_given_x
```

Having all of these probabilities, we can put them together and return the class that corresponds to the largest conditional probability:

$$ y = \operatorname*{argmax}_{i}\ \ P(y_i | \boldsymbol{x})$$

```python
probs = pd.Series({
    "Party": p_of_party_given_x, 
    "Study": p_of_study_given_x, 
    "TV": p_of_tv_given_x, 
    "Pub": p_of_pub_given_x})
probs
```

And here is finally the returned class (label)

```python
probs.idxmax()
```

## A simple implementation

Having seen how this classifier manually works, let's implement it.

```python
class NaiveBayesClassifier:
    def fit(self, X, y, features, target):
        """
        Sets things up for classification by preprocessing the training dataset
        and calculating the priors
        """
        self.X = X
        self.y = y
        self.features = features
        self.ds = my.to_dataframe(X, y, features, target)
        self.labels = self.ds.iloc[:, -1].unique()
        self.prior = self.ds.iloc[:, -1].value_counts() / self.ds.iloc[:, -1].count()
        
        return self
    
    def predict(self, X_unseen):
        """
        Returns the most probable label (or class) for each unseen input
        """
        if X_unseen.ndim == 1:
            x = pd.Series(X_unseen, index= self.features)
            prob = pd.Series(0.0, index= self.labels)
            for lbl in self.labels:
                df_per_lbl = self.ds.where(self.ds.iloc[:, -1] == lbl).dropna()

                likelihood = 1
                for f in self.features:
                    prob_per_lbl = df_per_lbl[f].value_counts() / df_per_lbl[f].count()
                    likelihood *= prob_per_lbl[x[f]] if x[f] in prob_per_lbl else 0

                prob[lbl] = likelihood * self.prior[lbl]
                
            return prob.idxmax()
        
        else:
            return np.array([self.predict(ex) for ex in X_unseen])
```

Make sure to study this implementation and relate it to the example above.

Let's test this implementation using the same above data, repeated here for convenience:

```python
df = pd.DataFrame({
    'deadline': ['Urgent', 'Urgent', 'Near', 'None', 'None', 'None', 'Near', 'Near', 'Near', 'Urgent'],
    'party': ['Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'No'],
    'lazy': ['Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No'],
    'y': ['Party', 'Study', 'Party', 'Party', 'Pub', 'Party', 'Study', 'TV', 'Party', 'Study']
})

df
```

```python
X, y, features, target = my.from_dataframe(df)
nb = NaiveBayesClassifier().fit(X, y, features, target)
print(nb.predict(np.array(['Near', 'No', 'No'])))
```

This is the same result we got from the detailed example above. Here is the training confusion matrix.

```python
cm = my.confusion_matrix(y, nb.predict(X))
accuracy = np.trace(cm) / np.sum(cm)

print(cm)
print('Training accuracy: ', accuracy)
```

## Gaussian naive Bayes classifier
The previous classifier does not work with continuous-valued features. For these features, we estimate the prior probabilities like before based on the training set. For the likelihood (conditional) probabilities, we make the assumption that they follow the normal (or Gaussian) distribution. Then we use training set to estimate the $\mu$ and $\sigma$ parameters of these Gaussian distributions.

```python
class GaussianNaiveBayesClassifier:
    def fit(self, X, y, features, target):
        """
        Sets things up for classification by preprocessing the training dataset
        and calculating the priors
        """
        self.X = X
        self.y = y
        self.features = features

        self.ds = my.to_dataframe(X, y, features, target)
        self.labels = self.ds.iloc[:, -1].unique()
        self.prior = self.ds.iloc[:, -1].value_counts() / self.ds.iloc[:, -1].count()
        self.mu = self.ds.groupby(target).mean()
        self.std = self.ds.groupby(target).std()
        
        return self
    
    def predict(self, X_unseen):
        """
        Returns the most probable label (or class) for each unseen input
        """
        if X_unseen.ndim == 1:
            x = pd.Series(X_unseen, index= self.features)
            prob = pd.Series(0.0, index= self.labels)
            for lbl in self.labels:
                df_per_lbl = self.ds.where(self.ds.iloc[:, -1] == lbl).dropna()
                likelihood = 1
                for f in self.features:
                    likelihood *= st.norm.pdf(x[f], loc= self.mu.at[lbl, f], scale=self.std.at[lbl, f])

                prob[lbl] = likelihood * self.prior[lbl]
                
            return prob.idxmax()
        
        else:
            return np.array([self.predict(ex) for ex in X_unseen])
```

### An example
First we need to generate some data. For visualization, we'll generate two-dimensional(x1 and x2 features) data with three classes: -1, 0, 1. Each class has 100 examples:

```python
basis = np.array([[3, 14],[4, 4],[10, 9]])
a_ = np.concatenate((
        basis[0,0] + np.random.normal(loc=1, scale=2.5, size=(100, 1)), 
        basis[0, 1] + np.random.normal(loc=1, scale=2.6, size=(100, 1)), 
        np.full(100, -1).reshape(100,1)), axis= 1)

b_ = np.concatenate((
        basis[1,0] + np.random.normal(loc=.8, scale=2.5, size=(100, 1)), 
        basis[1, 1] + np.random.normal(loc=.8, scale=2.4, size=(100, 1)), 
        np.full(100, 0).reshape(100,1)), axis= 1)

c_ = np.concatenate((
        basis[2,0] + np.random.normal(loc=.5, scale=2.5, size=(100, 1)), 
        basis[2, 1] + np.random.normal(loc=.5, scale=2.6, size=(100, 1)), 
        np.full(100, 1).reshape(100,1)), axis= 1)
data = pd.DataFrame(np.concatenate([a_, b_, c_], axis=0), columns=['x1', 'x2', 't'])
data
```

Next we split the data into training and testing sets:

```python
X, y, features, target = my.from_dataframe(data)
print(y.shape)
X_train, X_test, y_train, y_test = my.train_test_split(X, y.ravel(), test_size=.3)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Let's create and train a classifier.

```python
gnb = GaussianNaiveBayesClassifier().fit(X_train, y_train.ravel(), features, target)
```

Here is how this classifier performs on the training data.

```python
cm = my.confusion_matrix(y_train, gnb.predict(X_train))
accuracy = np.trace(cm) / np.sum(cm)

print(cm)
print('Training accuracy: ', accuracy)

my.plot_decision_regions(X_train, y_train, gnb, resolution=0.5, title="Training data")
```

And here is how it performs on the testing data.

```python
cm = my.confusion_matrix(y_test, gnb.predict(X_test))
accuracy = np.trace(cm) / np.sum(cm)

print(cm)
print('Training accuracy: ', accuracy)

my.plot_decision_regions(X_test, y_test, gnb, resolution=0.5, title="Testing data")
```

## Gaussian naive Bayes in `scikit-learn`

`scikit-learn` has a module dedicated to naive Bayes. Multiple naive Bayes classifiers are implemented in this module. The different naive Bayes classifiers differ mainly by the assumptions they make regarding the distribution of $P(\boldsymbol{x} | y_i)$. Here is an example using `scikit-learn` implementation of Guassian naive Bayes on the above dataset.

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB().fit(X_train, y_train)

fig = plt.figure(figsize=(16,5))
ax = fig.add_subplot(1,2,1)
my.plot_confusion_matrix_with_accuracy(gnb, X_train, y_train, ax=ax)
ax = fig.add_subplot(1,2,2)
my.plot_decision_regions(X_train, y_train, gnb, resolution=0.5, title="Training data", ax=ax)
```
