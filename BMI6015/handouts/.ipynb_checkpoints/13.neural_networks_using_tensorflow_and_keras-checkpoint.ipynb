## Introduction to Tensorflow

According to its [website](https://www.tensorflow.org), "TensorFlow is an end-to-end open source platform for machine learning". Together with Keras, they represent one of the most commonly used deep learning frameworks. This tutorial is meant to be a starting point and by no means is it meant be complete or thorough. Keep in mind that we are building on what we already know (NumPy, Pandas, and Sci-kit Learn) and not replacing it.

Tensorflow is much like NumPy with three major differences: 
* TensorFlow can leverage GPUs and TPUs.
* TensorFlow can automatically compute the gradients of arbitrary differentiable tensor operations.
* TensorFlow computation can be distributed to large numbers of devices on one or more machines.


A **tensor** is basically a multidimensional array. It generalizes scalars, vectors, and matrices. A tensor is defined by three key attributes:
* **Rank**: The number of axes or dimensions. You can find the rank of a tensor using the attribute `ndim`
* **Shape**: A tuple of integers describing how many units along each dimension. You can find the shape of a tensor using the attribute `shape` 
* **Data type**: The type of the data contained in the tensor. You can find the data type of a tensor using the attribute `dtype` 

Given this definition, both NumPy arrays and Tensorflow tensors are considered tensors. Let's see a couple of examples:

## Calculating gradients with GradientTape

One important reason why Tensorflow is used for deep learning is its ability to calculate the gradients (partial derivatives) of any differentiable function made of one or more tensor operations. This makes it easy to create gradient-based models such the ones that ones covered in this course. 

Let see an example.Here we have this function $$ y = f(x) = cos^3(x) + sin(x)$$

which looks like this.

```python
x = np.linspace(-4, 4, 100)
y = np.cos(x)**3 +  np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

We [know from calculus](https://www.wolframalpha.com/input?i=cos%28x%29%5E3+%2B+sin%28x%29) that the derivative of this function is 
$$ f' = cos(x) - 3 sin(x) cos^2(x)$$

This is called **symbolic differentiation** and is done either manually or using symbolic mathematical systems such Mathematica. Both of these options won't work for deep learning.

Tensorflow, along with other Deep Learning frameworks such as PyTorch, uses a technique called **revese-mode auto-differentiation** to calculates the gradients of a function with regard to its inputs. In Tensorflow this is done using the `tf.GradientTape` scope. Let's instruct Tensorflow to calculates the derivative of the above function. For this to work, we have to use variable tensors and tensor operations.

```python
x_t = tf.Variable(x, dtype=tf.float32)
with tf.GradientTape() as tape:
    y_t = tf.cos(x_t)**3 +  tf.sin(x_t)
    
grad = tape.gradient(y_t, x_t)
```

As you can see, we converted our NumPy `x` array into a variable tensor `x_t` and within the GradientTape scope, we calculated the function output tensor `y_t` using (and this is critical) tensor operations such as `tf.cos` and `tf.sin`. Having done that, we can get the gradient of `y_t` with respect to `x_t` by calling `tape.gradient(y_t, x_t)`. 

Let's plot this function along with the two derivatives we have for it: the one using symbolic differentiation and the one using Tensorflow's auto-differentiation.

```python
fig, ax = plt.subplots(1, 3, figsize=(16, 4))

y_prime = np.cos(x) - 3 * np.sin(x) * np.cos(x)**2
ax[0].plot(x, y, linewidth=2)
ax[0].set_ylabel('y')
ax[0].set_xlabel('x')
ax[0].set_title('Original function')
ax[0].grid()

ax[1].plot(x, y_prime, linewidth=2)
ax[1].set_ylabel('y\'')
ax[1].set_xlabel('x')
ax[1].set_title('Derivative - Manual differtiation')
ax[1].grid()

ax[2].set_title('Derivative - Auto-differtiation')
ax[2].set_ylabel('y\'')
ax[2].set_xlabel('x')
ax[2].plot(x, grad, linewidth=2)
ax[2].grid()
plt.show()
```

As you can see both derivatives are identical.

Notice that once `gradient()` is called, the tape is automatically deleted. If you need to call `gradient()` more than once, you must make the tap `persistent`. You must also delete the tape once you are done with it.

```python
x_t = tf.Variable(x, dtype=tf.float32)
with tf.GradientTape(persistent=True) as tape:
    y1 = tf.cos(x_t)**3 +  tf.sin(x_t)
    y2 = tf.cos(x_t) * tf.sin(x_t)
    
grad_1 = tape.gradient(y1, x_t)
grad_2 = tape.gradient(y2, x_t)
del tape

grad_1, grad_2
```

## Multi-layer Perceptron using Tensorflow

Now that we know how to use tensorflow, let's use it to implement a Multi-Layer Perceptron.

In this section we use pure Tensorflow to implement a simple MLP with one hidden layer like the figure below. As the name implies, a multilayer perceptron (MLP) is a set of perceptrons arranged into layers. For this implementation, we will be using the ReLU activation function (`tf.nn.relu`) for the hidden layer and the logistic (sigmoid) activation function (`tf.nn.sigmoid`) for the output layer. 

<img style="-webkit-user-select:none; display:block; margin:auto; padding:env(safe-area-inset-top) env(safe-area-inset-right) env(safe-area-inset-bottom) env(safe-area-inset-left);" src="">

Here is a simple class implementing a multi-layer perceptron classifier.

```python
class MLPClassifier:
    def __init__(self, nHidden, nOutput=1):
        self.nHidden = nHidden
        self.nOutput = nOutput
        self.W = []
    
    def MSE(self, actual, predicted):
        return tf.reduce_mean(tf.square(actual - predicted))
    
    def forward(self, X):
        X = tf.concat([tf.ones((len(X),1)), X],axis=1)
        
        """ Activations of hidden layer """
        self.a = tf.nn.relu(tf.matmul(X, self.W[0]));
        # Adding bias to the hidden layer
        self.a = tf.concat([tf.ones((len(self.a), 1)), self.a], axis=1)

        return tf.nn.sigmoid(tf.matmul(self.a, self.W[1]))
        
    def fit(self, X, y, epochs=200, eta=0.01):
        N, M = X.shape
        self.errors = []
        
        # STEP 1: Initialize parameters
        self.W.append(
            tf.Variable(initial_value=tf.random.normal([M + 1, self.nHidden], mean=.1, stddev=.2)))
        self.W.append(
            tf.Variable(initial_value=tf.random.normal([self.nHidden + 1, self.nOutput], mean=.1, stddev=.2)))
        
        history = { "losses": [] } # For tracking progress
        for n in range(epochs):
            with tf.GradientTape() as tape:
                # STEP 2: Calculate the predicted output 
                y_predicted = self.forward(X)
        
                # STEP 3: Calculate the loss function
                loss = self.MSE(y, y_predicted)
    
            # STEP 4: Calculate the gradients of the loss with respect the W parameters
            grad_W0, grad_W1 = tape.gradient(loss, [self.W[0], self.W[1]])

            # STEP 5: Update the W parameters.
            self.W[0].assign_sub(eta * grad_W0)
            self.W[1].assign_sub(eta * grad_W1)
        
            history["losses"].append(loss)
            
        return history
    def predict(self, X):
        return (self.forward(X)[:, 0] > 0.5).numpy().astype(int)
```

Let's test it using the following data:

```python
n_per_class = 300

# Class 0 data
X0 = np.random.multivariate_normal(
    mean=[1, 5],
    cov=[[1, 0.5],[0.5, 1]],
    size=n_per_class)

# Class 1 data
X1 = np.random.multivariate_normal(
    mean=[5, 3],
    cov=[[1, 0.6], [0.6, 1]],
    size=n_per_class)

# Putting the data together
X = tf.cast(tf.concat((X0, X1), axis=0), dtype=tf.float32)
y = tf.concat((tf.zeros((n_per_class, 1)), tf.ones((n_per_class, 1))), axis=0)
```

```python
mlp = MLPClassifier(16, 1)
history = mlp.fit(X, y, epochs=500)
```

And let's display the results:

```python
fig, ax = plt.subplots(1, 3, figsize=(16,4))

ax[0].scatter(X[:, 0], X[:, 1], c=y[:, 0])
ax[0].set_xlabel('$x_1$')
ax[0].set_ylabel('$x_2$')
ax[0].set_title('Original data')
ax[0].grid()

predictions = mlp.predict(X)
my.plot_decision_regions(X, y, mlp, ax=ax[1])

ax[1].set_title('Predictions')
ax[1].grid()

# Plot the decay of loss function
ax[2].plot(history["losses"])
ax[2].set_xlabel('epochs')
ax[2].set_ylabel('Losses')
ax[2].set_title('Loss functions')
ax[2].grid()

plt.show()
```

## Introduction to Keras

Keras is a powerful well-designed high-level framework built on top of Tensorflow. It makes creating and training neural networks super easy. It is packaged as part of Tensorflow under the `tf.keras` module. More information about Keras can be found at its [official web site](https://keras.io).

In this section, we'll use Keras to build and train neural networks that can be as deep as we like. We start by importing it:

```python
from tensorflow import keras
```

Keras operates on a principle called *the progressive disclosure of complexity*. It provides multiple ways to do the same thing. For straightforward deep neural networks (DNNs), the **Sequential model** is all you need. For those who want more control, the **Functional API** is provided. For ultimate control and/or if you want to implement DNNs in an object-oriented manner, you can subclass the **Layer** and **Model** classes. And, of course, you can mix and match between these three API's.

This notebooks covers Kera's Sequential Model. This is the simplest way to implement an ANN or DNN in Keras. It builds these models by stacking layers on top of one another. Creating and training models in Karas has a simple workflow:

* Create the model
* Configure the model for training by specifying the loss function, the optimizer, and the metric(s) to use to measure its performance. This step is done by calling the `compile` method of the model.
* Train the model by calling the `fit` method.
* Evaluate the model on the set-aside test data by calling the `evaluate` method.
* Make predictions, after that, using the `predict` method.

 Let's get started:

### Linear regression

We'll start by creating a synthetic dataset and splitting it into training and test sets.

```python
x = np.arange(0, 45, step=.1)
y = 3 * x + 2 + np.random.normal(loc=2, scale=10, size=len(x))

# shuffling
indexes = np.random.permutation(len(x))
x = x[indexes]
y = y[indexes]

# splitting
split_at = int(.8 * len(x))
x_train, x_test = tf.constant(x[0:split_at]), tf.constant(x[split_at:])
y_train, y_test = tf.constant(y[0:split_at]), tf.constant(y[split_at:])
```

Let's normalize the data. This is needed to make sure that all input features are of of similar scale. We'll use the training data for this. You should not use the test data in the calculation of the mean and standard deviation used for normalization. Doing so harms the validity of you model and causes the test data to leak.

As an exercise, try fitting this data without the following normalization step.

```python
mean = tf.reduce_mean(x_train)
std = tf.math.reduce_std(x_train)
x_train = (x_train - mean)/std
x_test = (x_test - mean)/std
```

We then define a sequential model with one dense (a fully connected hidden) layer. Because this is linear regression, we will leave this layer without an activation function, which effectively means that it's linear activation defined as:

$$g(\mathbf{x} \mathbf{w} + b) = \mathbf{x} \mathbf{w} + b$$


```python
model = keras.Sequential([keras.layers.Dense(1)])
```

This is univariate linear regression, so the number of units (neurons) in this single layer is 1. That is the first argument we passed to the `Dense` layer. The `Sequential` class takes a list of layers as an argument. This is where we specify the stack of layers that make up the model. In this example we have a single layer.

Every model has a name; if a name is not given, Keras generates one for it.

```python
model.name
```

We can provide our own name for the model using the `name=` parameter.

```python
model = keras.Sequential([keras.layers.Dense(1)], name="linear_regressor")
model.name
```

We can print this in a readable way by calling its `summary()` method. Doing so however will cause an error. Uncomment the code below to see the error.

```python
# model.summary()
```

This same will happen if you try to print the weight parameters of the model. Uncomment the code below to see the error.

```python
# model.weights
```

These errors are due to the fact that the number of weights is determined by the shape of the input data, and we have not told this model how many features or dimensions) that data will have. 

One way to pass that information to Keras is by passing the `input_shape=` argument to the dense layer like this:

```python
model = keras.Sequential([keras.layers.Dense(1, input_shape=(1,))], name="linear_regressor")
```

And by doing so, we can print the summary of this model:

```python
model.summary()
```

and its weights:

```python
model.weights
```

All of these weights of this model are trainable; they are learned from the data using the backpropagation algorithm.

```python
model.trainable_weights
```

Other models may have both trainable and non-trainable weights. Notice here a few things:
* We have two weight parameters: one for the intercept of the regression line and one for its slope. 
* Both of these parameters will be learned from the data when we train this model. Because of that, these parameters are called "Trainable params" by Keras.
* Looking at the values of these weights, you'll see that one of them (the intercept or bias) is initialized to 0.0 while the slope is initialized to a small random value. This corresponds to STEP 1 of the gradient descent.
* Both of these parameters are variable tensors. **Why?**

Another way of telling Keras about the dimensionality of the input data, is by having an `Input` layer: it is a specification of the input data and not a real layer. 

```python
model = keras.Sequential([
    keras.layers.Input(shape=(1,)),
    keras.layers.Dense(1)], name="linear_regressor")
model.summary()
```

Notice that the above model could be created by adding each layer programmatically one at a time using the `add()` method. This again shows how flexible Keras is.

```python
model = keras.Sequential(name="linear_regressor")
model.add(keras.layers.Input(shape=(1,)))
model.add(keras.layers.Dense(1, activation='linear'))
model.summary()
```

We can even plot this model:

```python
keras.utils.plot_model(model, show_shapes=True)
```

Before we can train this model, we need to configure it by specifying the following:
* What optimizer to use. There are many gradient-based optimizers to choose from. Here we'll use the SGD (stochastic  gradient descent). You can see the list of supported optimizers [at this page](https://keras.io/api/optimizers/).
* What loss function to use. A list of the supported losses is available [at this page](https://keras.io/api/losses/). This a regression problem, so we will be using the `mean_squared_error` loss.
* What metric or metrics use to judge the performance of the model. A list of supported metrics can be found [at this page](https://keras.io/api/metrics/). We will be using the `mean_absolute_error` metric.

We provide this information to the model using the `compile` method.

```python
model.compile(optimizer="SGD",
            loss='mean_squared_error',
            metrics=['mean_absolute_error'])
```

The `SGD`, `mean_squared_error`, and `mean_absolute_error` are shortcuts to the actual optimizer, loss and metrics to use. We can provide the above information to the model by calling the actual optimizer, loss, and metric classes. 

```python
model.compile(optimizer=keras.optimizers.SGD(),
              loss=keras.losses.MeanSquaredError(),
              metrics=[keras.metrics.MeanAbsoluteError()])
```

Using these classes give us more control. We can for instance provide our own value for the optimizer's learning rate:

```python
model.compile(optimizer=keras.optimizers.SGD(learning_rate=0.01),
              loss=keras.losses.MeanSquaredError(),
              metrics=[keras.metrics.MeanAbsoluteError()])
```

The compile method configures the model for training and creates the static computational graph needed to compute gradients and run backpropagation.

We can now train the model by calling its `fit` method. 

```python
epochs = 6
hist = model.fit(x_train, y_train, batch_size=8, epochs=epochs, validation_split=.2, verbose=True)
```

The results of this function is a history of what went on during training. The `validation_split=` argument instructs Keras to use 20% of the training dataset for validation. We are also using batches of size 8. Here are the parameters we pass to the `fit` method.

```python
hist.params
```

```python
print(hist.epoch)
```

The `fit()` method implements the training loop. A `verbose=False` argument can be used to make the training silent; otherwise a large printout will be displayed. Obviously, the loop is run for `epochs` times.

Let's print the trained weights:

### Logistic regression
The model above has one dense layer with no activation. If we add a sigmoid activation to this layer, we get a logistic regression model. The sigmoid function is defined as:

$$ \sigma(z) = \frac{1}{1 + e^{-z}}$$

We'll try this new model on the versicolor and virginica iris flowers. Let's fetch the data and split it into two sets: training and test.

```python
from sklearn.datasets import load_iris
iris = load_iris()

# Get only the versicolor and virginica flowers and change their classes from 1 and 2 to 0 and 1
X, y = iris.data[50:], iris.target[50:] - 1

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

The dataset has 4 input features. Using the sequential model, we write:

```python
model = keras.Sequential([
    keras.layers.Input(shape=(4,)),
    keras.layers.Dense(1, activation="sigmoid")
], name="logistic_regressor")

model. summary()
```

The activation function is specified as an argument to the dense layer. 

We then compile the model. Because this is a classification problem, we use `accuracy` as a metric and since this is binary classification, we'll use `binary_crossentropy` as a loss function.

```python
model.compile(optimizer=keras.optimizers.SGD(learning_rate=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

Let's train it:

```python
epochs = 40
hist = model.fit(X_train, y_train, batch_size=16, epochs=epochs, validation_split=.2, verbose=False)
```

And evaluate it.

```python
scores = model.evaluate(X_test, y_test, batch_size=8)
print("Test loss:", scores[0])
print("Test accuracy:", scores[1])
```

```python
results = pd.DataFrame(hist.history)
fig = plt.figure(figsize=(16, 4))

results[['loss', 'val_loss']].plot(ax=fig.add_subplot(1, 2, 1), grid=True)
results[['accuracy', 'val_accuracy']].plot(ax=fig.add_subplot(1, 2, 2), grid=True)
```

### Multi-layer Perceptron

As a final example, we'll implement a deeper yet simple MLP with multiple hidden layers and one output layer. Let's start with the data:

```python
n_per_class = 800

X = tf.cast(tf.concat((
    np.random.multivariate_normal(mean=[1, 5], cov=[[1.5, .6], [.6, 1]], size=n_per_class),
    np.random.multivariate_normal(mean=[7, 8], cov=[[2, 0.8], [0.8, 2]], size=n_per_class),
    np.random.multivariate_normal(mean=[10, 2], cov=[[2.5, 0.7], [0.7, 2.5]], size=n_per_class)
), axis=0), dtype=tf.float32)

y = tf.cast(
    tf.concat((tf.zeros((n_per_class, 1)), tf.ones((n_per_class, 1)), tf.fill((n_per_class, 1), 2.)), axis=0), 
    dtype=tf.float32)

X.shape, y.shape
```

Let's shuffle the data:

```python
D = tf.random.shuffle(tf.concat([X, y], axis=1))
X, y = D[:, :-1], tf.reshape(D[:, -1], shape=(-1, 1))
X.shape, y.shape
```

Let's split it into three sets: training, validation, and test.

```python
X_train, y_train = X[:1600,:], y[:1600, :] 
X_val, y_val = X[1600:2000,:], y[1600:2000, :] 
X_test, y_test = X[2000:,:], y[2000:, :] 
```

We can normalize the data here using means and standard deviations calculated using the training data. Or we can use normalization pre-processing layer of Keras. We'll do the latter.

Let's plot the training data:

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train[:, 0])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

We'll use a sequential model with 4 hidden dense layers with ReLU activation each with 8 units. We might not need all of these layers, but this gives us an example of a deep-ish network. Before these layers, we use a pre-processing layer to normalize the data based on the training set. Because this is a multiclass classifier, the output layer has 3 units (one for every class), and a softmax activation is used.

```python
mlp = keras.Sequential([
    keras.layers.Input(shape=(2,)),
    keras.layers.Normalization(mean=tf.reduce_mean(X_train, axis=0), 
                               variance=tf.math.reduce_variance(X_train, axis=0)),
    keras.layers.Dense(8, activation="relu"),
    keras.layers.Dense(8, activation="relu"),
    keras.layers.Dense(8, activation="relu"),
    keras.layers.Dense(8, activation="relu"),
    keras.layers.Dense(3, activation="softmax")
], name="mlp_classifier")
```

Let's summarize and plot this model:

```python
mlp.summary()
keras.utils.plot_model(mlp, show_shapes=True)
```

Let's compile this model. This is a multiclass classification with numerical values for classes. We'll use the `sparse_categorical_crossentropy` loss and the `accuracy` metric. We'll use `rmsprop` for optimizer.

```python
mlp.compile(optimizer="rmsprop",
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy'])
```

We not train it. We'll pass our validation set to the fit function using the `valiation_data=` argument.

```python
epochs = 10
hist = mlp.fit(X, y, batch_size=16, epochs=epochs, validation_data=(X_val, y_val), verbose=False)
```

Here are the results:

```python
fig = plt.figure(figsize=(16, 4))
results = pd.DataFrame(hist.history)
results[['loss', 'val_loss']].plot(ax=fig.add_subplot(1, 2, 1), grid=True)
results[['accuracy', 'val_accuracy']].plot(ax=fig.add_subplot(1, 2, 2), grid=True)
```

Let's evaluate it:

```python
scores = mlp.evaluate(X_test, y_test, batch_size=8)
print("Test loss:", scores[0])
print("Test accuracy:", scores[1])
```

Finally let's plot the decision regions.

```python
def plot_decision_regions(X, y, learner, resolution=0.1, title="Decision regions", ax=None):    
    if ax is None:
        fig = plt.figure(figsize=(16,9))
        ax = fig.add_subplot(1,1,1)
        
    # Plot the decision boundary.
    x_min, x_max = tf.math.reduce_min(X[:,0]) - 1, tf.math.reduce_max(X[:,0]) + 1
    y_min, y_max = tf.math.reduce_min(X[:,1]) - 1, tf.math.reduce_max(X[:,1]) + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, resolution),
                         np.arange(y_min, y_max, resolution))
    Z = np.argmax(learner.predict(np.array([xx.ravel(), yy.ravel()]).T), axis=1)

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)

    plt.rcParams['pcolor.shading'] ='nearest'
    ax.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired, alpha=.75)

    # Plot the points
    ax.scatter(
        x=X[:, 0],
        y=X[:, 1], c=y, cmap=plt.cm.Paired,
        edgecolor='k', s=60)
    
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())

    # plt.legend(loc='best')
    ax.set_title(title)
    
fig, ax = plt.subplots(1, 2, figsize=(16,4))

ax[0].scatter(X_test[:, 0], X_test[:, 1], c=y_test[:, 0])
ax[0].set_xlabel('$x_1$')
ax[0].set_ylabel('$x_2$')
ax[0].set_title('Original data')
ax[0].grid()

pp = np.argmax(mlp.predict(X_test), axis=1)
plot_decision_regions(X_test, pp, mlp, ax=ax[1])

ax[1].set_title('Predictions')
ax[1].grid()

plt.show()
```
