## Read & Clean up the Data

You've already done data cleanup in the previous homework, so we'll provide this for you. 

```python
avy_df = pd.read_csv('./avalanches.csv')

# Clean dates and elevation
avy_df['Date'] = pd.to_datetime(avy_df['Date'])
avy_df['Year'] = avy_df['Date'].dt.year.astype('Int64')
avy_df['Month'] = avy_df['Date'].dt.month.astype('Int64')
avy_df['Elevation_Feet'] = pd.to_numeric(avy_df['Elevation'].str.replace('\'', '').str.replace(',', ''))

# Filter out null dates and incomplete years
avy_df = avy_df[avy_df['Date']==avy_df['Date']]
avy_df = avy_df[avy_df['Year'] > 2009]
```

# Part 1: Violin Plot Per Region

* Create a horizontal violin plot of distributions of avalanche elevations by Region.  
* Sort the regions by the median avalanche elevation in a descending order. 
* Plot the mean on the violing plot. 
* Create lables for the regions that also show the number of avalances in the region. For example: "Salt Lake (3654)"
* Analyze the data. What are trends you see? 

Hints: 
* you'll need to create a series of elevation values for each region 
* Matplotlib violinplots don't like NAN values, make sure to drop them before you try to plot

You can choose your own aesthetics, but make sure it looks good. 

For your reference, here is our solution: 

![Example](violinplot.svg)

```python
df = avy_df[['Elevation_Feet', 'Region']].dropna(how='any')

median_elevations = df.groupby('Region')['Elevation_Feet'].median().sort_values(ascending=False)

# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Reorder regions based on median elevation (descending)
regions = median_elevations.index
region_stats = df.groupby('Region')['Elevation_Feet'].agg(['median', 'count']).sort_values(by='median', ascending=False)

# Create a list to store the elevation data for each region
elevation_data = [df[df['Region'] == region]['Elevation_Feet'] for region in regions]

# Create a horizontal violin plot
ax.violinplot(elevation_data, showmedians=True, showextrema=False, vert=False)

# Customize the plot
ax.set_yticks(range(1, len(regions) + 1))
ax.set_yticklabels([f'{region} ({count})' for region, count in zip(regions, region_stats['count'])])
ax.set_ylabel('Region')
ax.set_xlabel('Elevation (Feet)')
ax.set_title('Avalanche Elevation Distribution by Region')
```

### Your Analysis: 

First of all, there's a wide range of avalanches across regions, and doesn't seem to correlate with the Elevation distribution. In fact, the several features encoded (mean, median, avalanche distribution) doesn't seem to have much correlation together. The number of avalanches seem to influence how extreme/how many outliers the distributions have. For example, Salt Lake (3645 avalanches) have long tails. Abajos (14 avalanches) on the other hand, don't have very long tails.

# Part 2: Radial Bar Chart

* Create a radial bar chart of the average number of avalanches in a year by Aspect.
* Arrange the aspects in the usual convention (North at the top, West on the left, etc.).
* Color the barchart using a cyclic colormap (e.g., HSV).
* Provide an analysis.

Hints: 
* Here is some useful reading: [https://matplotlib.org/stable/api/projections/polar.html](https://matplotlib.org/stable/api/projections/polar.html)

Here is our solution: 

![A radial visualization of avalanches per aspect](radial.svg)

```python
fig, subfig = plt.subplots(subplot_kw={'projection':'polar'})

aspect_counts = avy_df.groupby(['Aspect', 'Year']).size().unstack(fill_value=0).T.mean()
aspect_counts = aspect_counts.reindex(index=['East', 'Northeast', 'North', 'Northwest', 'West', 'Southwest', 'South', 'Southeast'])

subfig.bar(
    height = aspect_counts.values,
    x = np.linspace(0.0, 2 * np.pi, len(aspect_counts), endpoint=False),
    width=[0.5 for x in range(len(aspect_counts ))],
    color = [plt.cm.hsv(30*x) for x in range(8)],
    alpha=0.5
    #bottom=100
)

subfig.set_ylim(0, 200)
subfig.set_rorigin(-150)

subfig.set_xticks(np.linspace(0.0, 2 * np.pi, len(aspect_counts), endpoint=False))
subfig.set_xticklabels(aspect_counts.index, size=8)

subfig.set_rlabel_position(207)
yticks = [0, 50, 100, 150]
subfig.set_yticks(yticks)
subfig.set_yticklabels([str(y) for y in yticks], fontsize=8)

subfig.set_title('Mean Annual Avalanches by Aspect')
```

### Your Analysis: 

There is a clear pattern here. The NE aspect has the greatest amount of avalanches, and has a ~linear decrease in avalanches both directions to SW having the least amount of avalanches. This visual makes it very easy to see the mean rates of avalanches by aspect. This would make it very easy to share and disseminate this information to hikers/skiiers, etc., promoting more safety.

# Part 3: Rolling average area subplots

* Create a chart where each subfigure is an area chart of the 7-day rolling average of the number of avalanches in a given ski season. Assume that a "ski season" starts July 1 and ends June 31.
* The subplots should overlap so that outliers are less of a problem.
* Provide an analysis of the data. 

Useful methods to consider: 
- Pandas's `.rolling()`
- matplotlib's `.fill_between()`, `.set_axis_off()`, `.tight_layout()`
- datetime's `.timetuple().tm_yday`

Make sure to check out the arguments of the above methods.

Here's how your result should look: 

![A series of line charts showing avalanches over time](line.svg)

```python
min_year = avy_df.Date.dt.year.min()
max_year = avy_df.Date.dt.year.max()

seasons = [avy_df[(avy_df.Date >= pd.Timestamp(year=target_year, month=7, day=1)) & (avy_df.Date <= pd.Timestamp(year=target_year + 1, month=6, day=30))] for target_year in range(min_year, max_year)]

seasons_upd = []

for season in seasons:
    curr_min_year = season.Date.dt.year.min()
    curr_max_year = season.Date.dt.year.max()
    first = season[season.Date.dt.year == curr_min_year]
    last = season[season.Date.dt.year == curr_max_year]
    
    dates = [pd.to_datetime(f'2011-{date.month}-{date.day}', format='%Y-%m-%d') for date in first.Date]
    dates.extend([pd.to_datetime(f'2012-{date.month}-{date.day}', format='%Y-%m-%d') for date in last.Date])
    
    dates.sort()
    
    curr_season = pd.DataFrame(sorted(dates), columns=['Date'])
    curr_season = curr_season.sort_index()
    seasons_upd.append(curr_season)
    
counts = [pd.Series(season.value_counts('Date')) for season in seasons_upd]
counts_sorted = [count.sort_index() for count in counts]
rolling_avgs = [count.rolling(window=7).mean() for count in counts_sorted]
```

```python
import matplotlib.dates as mdates

fig, subfig = plt.subplots(len(rolling_avgs), 1, sharex=True, sharey=True, figsize=(2,5))

for season in range(len(rolling_avgs)):
    subfig[season].fill_between(rolling_avgs[season].index, .5, rolling_avgs[season].values, alpha = (season+3)/(len(rolling_avgs)+6))
    subfig[season].set_title(f'{season+2010}/{season+2010+1}', loc='right', y=-.01, ha='left', size=12)
    if season != len(rolling_avgs)-1:
        subfig[season].set_axis_off()
    else:
        subfig[season].spines[['right', 'top', 'left']].set_visible(False)
        subfig[season].set_yticks([])
        subfig[season].patch.set_alpha(0.0)

fig.gca().xaxis.set_major_locator(mdates.MonthLocator())
fig.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b'))
fig.tight_layout(pad=-8)
fig.suptitle('7 Day Rolling Average of Avalanches by Season', y=1.05, x=0.75)
```
